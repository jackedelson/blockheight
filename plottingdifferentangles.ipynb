```python
# Plotting Trajectories from 25 to 65 degrees in 5 degree increments

import numpy as np 
import matplotlib.pyplot as plt 

v = 30
g = 9.80

theta = np.arange(25,65,5)[None,:]/180.0*(np.pi)

plt.figure()

tmax = ((2 * v) * np.sin(theta)) / g
vecT = tmax*np.linspace(0,1,100)[:,None]

x = ((v * vecT) * np.cos(theta))
y = ((v * vecT) * np.sin(theta)) - ((0.5 * g) * (vecT ** 2))

plt.plot(x,y)
plt.ylim([0,40])
plt.xlabel('Horizontal Displacement')
plt.ylabel('Vertical Displacement')
plt.title('Projectile Motion at Different Angles')
plt.show()
```

```python
# Plotting from 41 to 49 degrees in 2 degree increments

import numpy as np 
import matplotlib.pyplot as plt 

v = 30
g = 9.80

theta = np.arange(41,49,2)[None,:]/180.0*(np.pi)

plt.figure()

tmax = ((2 * v) * np.sin(theta)) / g
vecT = tmax*np.linspace(0,1,100)[:,None]

x = ((v * vecT) * np.cos(theta))
y = ((v * vecT) * np.sin(theta)) - ((0.5 * g) * (vecT ** 2))

plt.plot(x,y)
plt.ylim([0,28])
plt.xlim([80,100]) # cLoseup snapshot the projectile landings
plt.xlabel('Horizontal Displacement')
plt.ylabel('Vertical Displacement')
plt.title('Projectile Motion at Different Angles')
plt.show()

# as we can see, the green line hits the furthest which represents 45 degrees, which is what we expected
```
