```python
# Plotting Trajectories from 25 to 65 degrees in 5 degree increments

import numpy as np 
import matplotlib.pyplot as plt 

v = 7.6
g = 9.80

theta = np.arange(12,20,2)[None,:]/180.0*(np.pi)

plt.figure()

h = 0.0
tmax = h + ((2 * v) * np.sin(theta)) / g
vecT = tmax*np.linspace(0,1,100)[:,None]

for h in np.arange(0.0,5.0,0.76):
    tmax = h + ((2 * v) * np.sin(theta)) / g
    print(tmax)
    


x = ((v * vecT) * np.cos(theta))
y = ((v * vecT) * np.sin(theta)) - ((0.5 * g) * (vecT ** 2))

plt.plot(x,y)
plt.ylim([0,1])
plt.xlim([0,4])
plt.xlabel('Horizontal Displacement')
plt.ylabel('Vertical Displacement')
plt.title('Projectile Motion at Different Angles')
plt.show()
```
