```python
# Hit run to start the simulator. 
# Then enter you desired block height and x and y limits for the graph of the swimmer's trajectory
# All units are in SI

import numpy as np 
import matplotlib.pyplot as plt 

v = 7.8
g = 9.80

theta = 12/180*(np.pi)

plt.figure()

h = float(input("Enter the height of the block in meters: "))
tmax = h + ((2 * v) * np.sin(theta)) / g
vecT = tmax*np.linspace(0,1,100)[:,None]

x = ((v * vecT) * np.cos(theta))
y = h + ((v * vecT) * np.sin(theta)) - ((0.5 * g) * (vecT ** 2))
plt.plot(x,y)
plt.ylim([0,float(input("Enter y limit in trajectory display: "))])
plt.xlim([0,float(input("Enter x limit in trajectory display: "))])
plt.xlabel('Horizontal Displacement')
plt.ylabel('Vertical Displacement')
plt.title('Choose your block height simulation')
plt.show()

if h >= 58.2:
    print("YOUR SWIMMER WILL SUSTAIN FRACTURES IN THEIR THORACIC VERTABRAE.")
```
