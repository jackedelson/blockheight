```python
import numpy as np 
import matplotlib.pyplot as plt 

v = 7.8
g = 9.80

theta = 12/180*(np.pi)

plt.figure()

h = 0.76
tmax = h + ((2 * v) * np.sin(theta)) / g
vecT = tmax*np.linspace(0,1,100)[:,None]


x = ((v * vecT) * np.cos(theta))
y = h + ((v * vecT) * np.sin(theta)) - ((0.5 * g) * (vecT ** 2))
plt.plot(x,y)
plt.ylim([0,4])
plt.xlim([0,5])
plt.xlabel('Horizontal Displacement')
plt.ylabel('Vertical Displacement')
plt.title('Caeleb Dressel Start Simulation')
plt.show()
```
